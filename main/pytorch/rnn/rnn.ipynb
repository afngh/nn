```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import torch.optim as optim
from rich.progress import track
import math
```

```python
X = torch.tensor([
    [[2.], [1.], [5.]],
    [[4.], [4.], [2.]],
    [[7.], [0.], [3.]],
    [[1.], [8.], [2.]],
    [[5.], [5.], [5.]],
    [[9.], [1.], [1.]],
    [[3.], [6.], [2.]],
    [[0.], [4.], [7.]],
    [[8.], [3.], [4.]],
    [[2.], [2.], [6.]],
    [[6.], [5.], [1.]],
    [[1.], [9.], [4.]],
    [[4.], [7.], [3.]],
    [[5.], [2.], [8.]],
    [[3.], [3.], [9.]],
    [[100.],[20.],[120.],]
], dtype=torch.float32)

y = torch.tensor([
    [8.],
    [10.],
    [10.],
    [11.],
    [15.],
    [11.],
    [11.],
    [11.],
    [15.],
    [10.],
    [12.],
    [14.],
    [14.],
    [15.],
    [15.],
    [240.],
], dtype=torch.float32)
```

```python
class RNN(nn.Module):
  def __init__(self):
    super().__init__()

    self.rnn = nn.RNN(
        input_size=1,
        hidden_size=8,
        batch_first=True
    )

    self.fc = nn.Linear(8,1)
  def forward(self,x):
    out, hidden = self.rnn(x)

    out = out[:,-1,:]

    out = self.fc(out)

    return out
```

```python
model = RNN()

loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(),lr=0.01)
```

```python
for epoch in track(range(5000),description="Training RNN"):
  yp = model.forward(X)

  loss = loss_fn(yp,y)
  optimizer.zero_grad()

  loss.backward()
  optimizer.step()
  if epoch % 1000 == 0:
    if math.ceil(loss.item()) == 0:
      break
    print(f"Epoch: {epoch} Loss: {loss.item()}")
```
